## Find Duplicates in an Array

```python
Input:  [4,3,2,7,8,2,3,1]
Output: [2,3]
```

```python
def findDuplicatesInArray(nums):
    nums.sort()
    duplicates = []

    for i in range(1, len(nums) - 1):
        if nums[i] == nums[i - 1]:
            if nums[i] not in duplicates:
                duplicates.append(nums[i])

    return duplicates


nums = [4, 3, 2, 1, 8, 9, 2, 3]
result = findDuplicatesInArray(nums)
print(result)
```

```python
def findDuplicatesInArray2(nums):
    seen = set()
    duplicates = []
    
    for num in nums:
        if num in seen:
            duplicates.append(num)
        else:
            seen.add(num)

    return duplicates

nums = [4, 3, 2, 1, 8, 9, 2, 3]
result = findDuplicatesInArray2(nums)
print(result)
```

# Product of Array Except Self


Given an array nums, return an array answer such that:
answer[i] = product of all elements of nums except nums[i]

Input: nums = [1,2,3,4]
Output: [24,12,8,6]


```python
def productOfArrayExceptSelf(nums):
    n = len(nums)
    result = [1] * n

    prefix = 1
    for i in range(n):
        result[i] = prefix
        prefix *= nums[i]

    suffix = 1
    for i in range(n - 1, -1, -1):
        result[i] * suffix
        suffix *= nums[i]

    return result

```
